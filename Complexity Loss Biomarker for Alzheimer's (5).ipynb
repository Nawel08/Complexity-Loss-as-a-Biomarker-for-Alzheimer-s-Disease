```python
def to_uint8(img):
  """Convertit une image en uint8"""
  img = np.nan_to_num(img)
  mn, mx = img.min(), img.max()
  if mx == mn:
      return np.zeros_like(img, dtype=np.uint8)
  scaled = 255 * (img - mn) / (mx - mn)
  return scaled.astype(np.uint8)
```

## 1 - Mesure de complexité

```python
def kolmogorov_complexity(data, method='gzip'):
    """Approximation K-complexity via compression"""
    if isinstance(data, np.ndarray):
        data = data.tobytes()

    original_size = len(data)
    if original_size == 0:
        return 0.0

    if method == 'gzip':
        compressed = gzip.compress(data, compresslevel=9)
    elif method == 'bz2':
        compressed = bz2.compress(data, compresslevel=9)
    elif method == 'lzma':
        compressed = lzma.compress(data, preset=9)
    else:
        raise ValueError(f"Unknown method: {method}")

    compressed_size = len(compressed)
    return compressed_size / original_size


def conditional_complexity(image1, image2, method='gzip'):
    """Distance informationnelle entre deux images"""
    if isinstance(image1, np.ndarray):
        image1 = image1.tobytes()
    if isinstance(image2, np.ndarray):
        image2 = image2.tobytes()

    cx = len(gzip.compress(image1))
    cy = len(gzip.compress(image2))
    cxy = len(gzip.compress(image1 + image2))

    nid = (cxy - min(cx, cy)) / max(cx, cy)
    return nid


def regional_complexity_map(image, grid_size=8):
    """Carte de complexité régionale"""
    h, w = image.shape
    step_h = h // grid_size
    step_w = w // grid_size

    complexity_map = np.zeros((grid_size, grid_size))

    for i in range(grid_size):
        for j in range(grid_size):
            region = image[i*step_h:(i+1)*step_h, j*step_w:(j+1)*step_w]
            complexity_map[i, j] = kolmogorov_complexity(region)

    return complexity_map

def conditional_kx_given_y(x, y, method="gzip"):
    x_bytes = x.tobytes()
    y_bytes = y.tobytes()

    if method == "gzip":
        Cy = len(gzip.compress(y_bytes))
        Cyx = len(gzip.compress(y_bytes + x_bytes))
    else:
        raise NotImplementedError()

    return Cyx - Cy
```

```python
def shannon_entropy(img, bins=256):
  """Entropie de Shannon"""
  h, _ = np.histogram(img.ravel(), bins=bins)
  p = h / h.sum()
  p = p[p>0]
  return -np.sum(p * np.log2(p))
```

```python
def resize_image(image, scale):
    """Redimensionne une image par un facteur"""
    h, w = image.shape
    new_h, new_w = int(h * scale), int(w * scale)
    img_pil = Image.fromarray(image)
    img_resized = img_pil.resize((new_w, new_h), Image.LANCZOS)
    return np.array(img_resized)


def extract_patches(image, size=32):
    """Extrait des patches non-overlapping"""
    h, w = image.shape
    patches = []
    for i in range(0, h - size + 1, size):
        for j in range(0, w - size + 1, size):
            patches.append(image[i:i+size, j:j+size])
    return patches


def compress_ratio(data, method='gzip'):
    """Wrapper pour kolmogorov_complexity"""
    return kolmogorov_complexity(data, method)
```

```python
def explore_dataset_structure(dataset_path):
    """Explore la structure du dataset"""
    dataset_path = Path(dataset_path)

    def print_tree(directory, prefix="", max_depth=3, current_depth=0):
        if current_depth > max_depth:
            return

        items = sorted(directory.iterdir())
        dirs = [item for item in items if item.is_dir()]
        files = [item for item in items if item.is_file()]

        for i, subdir in enumerate(dirs):
            is_last = (i == len(dirs) - 1) and len(files) == 0
            print(f"{prefix}{'└── ' if is_last else '├── '} {subdir.name}")

            img_count = len(list(subdir.glob('*.jpg'))) + len(list(subdir.glob('*.png')))
            if img_count > 0:
                print(f"{prefix}{'    ' if is_last else '│   '}    ({img_count} images)")

            new_prefix = prefix + ("    " if is_last else "│   ")
            print_tree(subdir, new_prefix, max_depth, current_depth + 1)

    print_tree(dataset_path)
    print()


def load_mri_dataset_from_kaggle():
    """Télécharge et charge le dataset"""
    path = kagglehub.dataset_download("lukechugh/best-alzheimer-mri-dataset-99-accuracy")
    print(f" Dataset downloaded to: {path}\n")

    return load_mri_dataset_from_path(path)


def load_mri_dataset_from_path(dataset_path):
    """Charge le dataset depuis un path"""
    dataset_path = Path(dataset_path)
    explore_dataset_structure(dataset_path)

    category_keywords = {
        'no': ('No Impairment', 0),
        'very mild': ('Very Mild Impairment', 1),
        'mild': ('Mild Impairment', 2),
        'moderate': ('Moderate Impairment', 3)
    }

    label_names = ['No Impairment', 'Very Mild Impairment',
                   'Mild Impairment', 'Moderate Impairment']

    images = []
    labels = []


    def scan_directory(directory, depth=0):
        if depth > 5:
            return

        for item in directory.iterdir():
            if item.is_dir():
                folder_name_lower = item.name.lower()

                matched = False
                for keyword, (category_name, label) in category_keywords.items():
                    if keyword in folder_name_lower:
                        img_files = list(item.glob('*.jpg')) + list(item.glob('*.png'))

                        for img_path in img_files:
                            try:
                                img = Image.open(img_path).convert('L')
                                img = img.resize((128, 128))
                                images.append(np.array(img))
                                labels.append(label)
                            except Exception as e:
                                print(f" Error: {img_path.name}")

                        if len(img_files) > 0:
                            print(f"   {category_name}: {len(img_files)} images (from {item.name})")
                        matched = True
                        break

                if not matched:
                    scan_directory(item, depth + 1)

    scan_directory(dataset_path)

    if len(images) == 0:
        print("\n No images found!")
        return np.array([]), np.array([]), label_names

    return np.array(images), np.array(labels), label_names
```

## 3 - Extraction des features


```python
def extract_rich_features(image):
    """Features enrichies - VERSION CORRIGÉE"""
    features = []

    # 1. Multi-scale compression
    for scale in [0.25, 0.5, 1.0]:
        resized = resize_image(image, scale)
        features.append(compress_ratio(resized, 'gzip'))

    # 2. Régions anatomiques approximatives
    h, w = image.shape
    hippocampus = image[int(h*0.5):int(h*0.8), int(w*0.3):int(w*0.7)]
    cortex = image[int(h*0.15):int(h*0.5), :]
    features.append(compress_ratio(hippocampus, 'gzip'))
    features.append(compress_ratio(cortex, 'gzip'))

    # 3. Variabilité locale
    patches = extract_patches(image, size=32)
    if len(patches) > 0:
        patch_complexities = [compress_ratio(p, 'gzip') for p in patches]
        features.append(np.std(patch_complexities))
        features.append(np.mean(patch_complexities))
    else:
        features.extend([0, 0])

    # 4. Complexité des gradients
    edges = sobel(image.astype(float))
    features.append(compress_ratio(edges.astype(np.uint8), 'gzip'))

    #shannon
    features.append(shannon_entropy(image))

    # 5. Multi-compressor ensemble
    features.append(compress_ratio(image, 'gzip'))
    features.append(compress_ratio(image, 'bz2'))
    features.append(compress_ratio(image, 'lzma'))

    return np.array(features)


def extract_complexity_features(images, verbose=True):
    """Extrait features pour tous les images"""
    features = []
    n = len(images)

    for idx, img in enumerate(images):
        if verbose and (idx + 1) % 100 == 0:
            print(f"   Processing: {idx+1}/{n}")
        features.append(extract_rich_features(img))

    return np.array(features)
```

```python

def mdl_classifier(X_train, y_train, X_test, verbose=True):
    """Classifieur MDL basique"""
    classes = np.unique(y_train)
    prototypes = {}

    for c in classes:
        class_images = X_train[y_train == c]
        prototypes[c] = np.mean(class_images, axis=0).astype(np.uint8)
        print(f"   Class {c}: {len(class_images)} samples")

    predictions = []
    n = len(X_test)

    for idx, test_img in enumerate(X_test):
        if verbose and (idx + 1) % 50 == 0:
            print(f"   {idx+1}/{n}")

        min_complexity = float('inf')
        predicted_class = None

        for c, prototype in prototypes.items():
            complexity = conditional_complexity(test_img, prototype)

            if complexity < min_complexity:
                min_complexity = complexity
                predicted_class = c

        predictions.append(predicted_class)

    return np.array(predictions)
```

# - Classification CRT

```python
def crt_classifier(images_train, labels_train, images_test):
    # complexity globale moyenne pour chaque classe
    cr0 = np.mean([ compress_ratio(img) for img, label in zip(images_train, labels_train) if label==0])
    cr1 = np.mean([compress_ratio(img) for img, label in zip(images_train, labels_train) if label==1])
    threshold = (cr0 + cr1) / 2
    # prédiction
    preds = []
    for img in images_test:
        cr = compress_ratio(img)
        preds.append(1 if cr < threshold else 0)
    return np.array(preds)
```

# - Classification BNN

```python
def bnn_classifier(images_train, labels_train, images_test, method='gzip'):
    # --- Precompute compressed versions of training images ---
    compressed_train = [gzip.compress(img.tobytes()) for img in images_train]
    sizes_train = [len(c) for c in compressed_train]

    preds = []

    for img_test in images_test:
        # Precompute for the test image
        img_test_bytes = img_test.tobytes()
        comp_test = gzip.compress(img_test_bytes)
        size_test = len(comp_test)

        min_dist = float("inf")
        best_label = None

        # Compare to *cached* compressed images
        for comp_train, size_train, img_train, label in zip(
            compressed_train, sizes_train, images_train, labels_train
        ):
            cxy = len(gzip.compress(img_test_bytes + img_train.tobytes()))
            nid = (cxy - min(size_test, size_train)) / max(size_test, size_train)

            if nid < min_dist:
                min_dist = nid
                best_label = label

        preds.append(best_label)

    return np.array(preds, dtype=int)


```

# - Classification ckc

```python
def ckc_classifier(train_images, train_labels, test_images):
    classes = np.unique(train_labels)
    pred = []

    for img_test in test_images:
        scores = {}
        for c in classes:
            imgs_c = [train_images[i] for i in range(len(train_images)) if train_labels[i] == c]
            vals = [conditional_kx_given_y(img_test, t) for t in imgs_c]
            scores[c] = np.mean(vals)

        pred.append(min(scores, key=scores.get))

    return np.array(pred)
```

```python
def mdl_classifier_features(X_train, y_train, X_test, verbose=True):
    """MDL classifier adapté aux vecteurs de features"""
    classes = np.unique(y_train)
    prototypes = {c: np.mean(X_train[y_train==c], axis=0) for c in classes}

    predictions = []
    n = len(X_test)

    for idx, x in enumerate(X_test):
        if verbose and (idx+1) % 50 == 0:
            print(f"   {idx+1}/{n}")
        min_dist = float('inf')
        pred_class = None
        for c, proto in prototypes.items():
            dist = np.linalg.norm(x - proto)  # distance vectorielle
            if dist < min_dist:
                min_dist = dist
                pred_class = c
        predictions.append(pred_class)

    return np.array(predictions)
```

```python
def bnn_classifier_features(X_train_features, y_train, X_test_features):
    preds = []
    for x in X_test_features:
        dists = np.linalg.norm(X_train_features - x, axis=1)
        idx_min = np.argmin(dists)
        preds.append(y_train[idx_min])
    return np.array(preds)
```

# Classification CKC

```python
def ckc_features(x, y):
    """Approximation CKC sur vecteurs de features"""
    # distance normalisée entre vecteurs
    return np.linalg.norm(x - y) / (np.linalg.norm(x) + np.linalg.norm(y) + 1e-8)
```

```python
from scipy.stats import kruskal
import numpy as np

def test_global_kruskal(complexities, labels, label_names):
    """
    Performs the global Kruskal–Wallis test across all Alzheimer stages.
    Returns the p-value and prints the test summary.
    """

    groups = []
    for i, name in enumerate(label_names):
        groups.append(complexities[labels == i])

    stat, p = kruskal(*groups)

    print("\n=== Global Kruskal–Wallis Test ===")
    print(f"H-statistic = {stat:.4f}, p-value = {p:.6f}")
    if p < 0.05:
        print("→ Significant differences detected between groups.")
    else:
        print("→ No significant overall difference between groups.")

    return p
```

```python
from statsmodels.stats.multitest import multipletests

def test_pairwise_kruskal(complexities, labels, label_names):
    """
    Performs pairwise Kruskal tests between all label pairs,
    with FDR-BH correction.
    Returns a structured list of results.
    """

    print("\n=== Pairwise Kruskal Tests (FDR corrected) ===")

    pairs = []
    p_values = []

    # Compare all pairs
    for i in range(len(label_names)):
        for j in range(i + 1, len(label_names)):
            group1 = complexities[labels == i]
            group2 = complexities[labels == j]

            stat, p = kruskal(group1, group2)

            pairs.append((i, j))
            p_values.append(p)

    # FDR correction
    rejected, corrected_p, _, _ = multipletests(p_values, method='fdr_bh')

    # Print results
    for k, ((i, j), corr_p, reject) in enumerate(zip(pairs, corrected_p, rejected)):
        decision = "SIGNIFICANT" if reject else "n.s."
        print(f"{label_names[i]} vs {label_names[j]}: p_corrected={corr_p:.6f} → {decision}")

    return pairs, corrected_p, rejected
```

```python
import numpy as np

def cohens_d(x, y):
    """Computes Cohen’s d between two distributions."""
    nx, ny = len(x), len(y)
    pooled_std = np.sqrt(((nx - 1)*np.var(x) + (ny - 1)*np.var(y)) / (nx + ny - 2))
    return (np.mean(x) - np.mean(y)) / pooled_std

def compute_effect_sizes(complexities, labels, label_names):
    """
    Computes Cohen’s d for all pairwise group comparisons.
    """

    print("\n=== Pairwise Effect Sizes (Cohen’s d) ===")

    effects = {}

    for i in range(len(label_names)):
        for j in range(i + 1, len(label_names)):
            group1 = complexities[labels == i]
            group2 = complexities[labels == j]

            d = np.abs(cohens_d(group1, group2))  # absolute value => magnitude only
            effects[(i, j)] = d

            magnitude = (
                "small" if d < 0.2 else
                "medium" if d < 0.5 else
                "large" if d < 0.8 else
                "very large"
            )

            print(f"{label_names[i]} vs {label_names[j]}: d = {d:.3f} → {magnitude}")

    return effects
```

```python
def run_full_complexity_analysis(complexities, labels, label_names):
    test_global_kruskal(complexities, labels, label_names)
    test_pairwise_kruskal(complexities, labels, label_names)
    compute_effect_sizes(complexities, labels, label_names)
```

```python
import numpy as np
import gzip, bz2, lzma
from scipy.stats import kruskal, mannwhitneyu, spearmanr
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.feature_extraction.image import extract_patches_2d, reconstruct_from_patches_2d
from itertools import combinations
```

Compression

```python
def compress_len_bytes(b, method='gzip'):
    if method=='gzip':
        return len(gzip.compress(b))
    if method=='bz2':
        return len(bz2.compress(b))
    if method=='lzma':
        return len(lzma.compress(b))
    raise ValueError(method)

def normalize_to_uint8(img, clip_percentiles=(1,99)):
    # scale an image to uint8 robustly (clip to percentiles to avoid outliers)
    lo, hi = np.percentile(img, clip_percentiles)
    imgc = np.clip(img, lo, hi)
    if hi == lo:
        return np.zeros_like(imgc, dtype=np.uint8)
    scaled = ((imgc - lo) / (hi - lo) * 255.0).astype(np.uint8)
    return scaled

def complexity_by_compressors(img, methods=('gzip','bz2','lzma')):
    """Return dict of normalized compressed sizes (len(compressed)/len(bytes))."""
    u8 = normalize_to_uint8(img)
    b = u8.tobytes()
    out = {}
    for m in methods:
        c = compress_len_bytes(b, method=m)
        out[m] = c / max(len(b), 1)
    return out
```

Template residual

```python
def build_template(controls):
    """controls: list/array of images (2D numpy). returns mean template (float)."""
    stack = np.stack(controls, axis=0).astype(np.float32)
    return stack.mean(axis=0)

def residuals_from_template(images, template):
    """images: list/array. returns list of residual floats (image - template)."""
    res = [img.astype(np.float32) - template.astype(np.float32) for img in images]
    return res
```

Predictive pca

```python
def fit_patch_pca_on_controls(controls, patch_size=(16,16), max_patches_per_img=200, n_components=50, random_state=0):
    """Extract patches from controls, fit PCA, return PCA model and patch_size."""
    all_patches = []
    for img in controls:
        # extract random patches
        patches = extract_patches_2d(img.astype(np.float32), patch_size, max_patches=max_patches_per_img, random_state=random_state)
        all_patches.append(patches.reshape(patches.shape[0], -1))
    if len(all_patches)==0:
        raise ValueError("No patches extracted")
    all_patches = np.vstack(all_patches)
    pca = PCA(n_components=min(n_components, all_patches.shape[1]), svd_solver='randomized', random_state=random_state)
    pca.fit(all_patches)
    return pca, patch_size

def reconstruct_image_with_patch_pca(img, pca, patch_size=(16,16), stride=None):
    """Patchwise reconstruction with PCA. stride default = patch_size (non-overlapping)."""
    if stride is None:
        stride = patch_size
    ph, pw = patch_size
    h, w = img.shape
    # Extract non-overlapping grid patches for speed
    patches = []
    coords = []
    for i in range(0, h - ph + 1, stride[0]):
        for j in range(0, w - pw + 1, stride[1]):
            p = img[i:i+ph, j:j+pw].ravel()
            patches.append(p)
            coords.append((i,j))
    patches = np.array(patches)
    if patches.size==0:
        # fallback: single flat reconstruction
        flat = img.ravel()[None,:]
        recon_flat = pca.inverse_transform(pca.transform(flat))[0]
        return recon_flat.reshape(img.shape)
    # project and reconstruct
    proj = pca.transform(patches)
    recon = pca.inverse_transform(proj)
    # assemble
    recon_img = np.zeros_like(img, dtype=np.float32)
    counts = np.zeros_like(img, dtype=np.float32)
    for (i,j), r in zip(coords, recon):
        recon_img[i:i+ph, j:j+pw] += r.reshape(ph,pw)
        counts[i:i+ph, j:j+pw] += 1.0
    counts[counts==0] = 1.0
    recon_img /= counts
    return recon_img

def residuals_from_pca(images, pca, patch_size=(16,16), stride=None):
    res = []
    for img in images:
        recon = reconstruct_image_with_patch_pca(img.astype(np.float32), pca, patch_size, stride)
        res.append(img.astype(np.float32) - recon.astype(np.float32))
    return res
```

Complexity computation

```python
def compute_dataset_complexities(images, methods=('gzip','bz2','lzma'), verbose=True):
    """Return dict of arrays: for each method an array of normalized complexities."""
    results = {m: [] for m in methods}
    for i, img in enumerate(images):
        if verbose and (i+1)%100==0:
            print(f"  computed {i+1}/{len(images)}")
        vals = complexity_by_compressors(img, methods=methods)
        for m in methods:
            results[m].append(vals[m])
    for m in methods:
        results[m] = np.array(results[m])
    return results
```

Statistical test

```python
def pairwise_tests_and_effects(values, labels, label_names, alpha=0.05):
    """values: 1d array, labels: ints. returns pairwise pvals corrected and effect sizes"""
    pairs = []
    pvals = []
    for (i,j) in combinations(range(len(label_names)),2):
        a = values[labels==i]
        b = values[labels==j]
        # use Mann-Whitney (equivalent to Kruskal pairwise)
        stat, p = mannwhitneyu(a, b, alternative='two-sided')
        pairs.append((i,j))
        pvals.append(p)
    rej, pcorr, _, _ = multipletests(pvals, method='fdr_bh', alpha=alpha)
    # effect sizes Cohen's d
    def cohens_d(x,y):
        nx,ny = len(x), len(y)
        sx = x.std(ddof=1); sy = y.std(ddof=1)
        pooled = np.sqrt(((nx-1)*sx* sx + (ny-1)*sy*sy)/(nx+ny-2))
        return (x.mean()-y.mean())/pooled
    effects = []
    for (i,j) in pairs:
        d = abs(cohens_d(values[labels==i], values[labels==j]))
        effects.append(d)
    return pairs, pvals, pcorr, rej, effects

def permutation_test_mean_diff(values, labels, group_a=0, group_b=1, n_perm=5000, seed=0):
    """Permutation test for difference in means between two groups"""
    rng = np.random.RandomState(seed)
    a = values[labels==group_a]
    b = values[labels==group_b]
    obs = a.mean() - b.mean()
    pooled = np.concatenate([a,b])
    n = len(a)
    perm_diffs = []
    for _ in range(n_perm):
        rng.shuffle(pooled)
        perm_diffs.append(pooled[:n].mean() - pooled[n:].mean())
    perm_diffs = np.array(perm_diffs)
    p = (np.sum(np.abs(perm_diffs) >= abs(obs)) + 1) / (n_perm + 1)
    return obs, p, perm_diffs

def plot_group_box(values, labels, label_names, title=''):
    plt.figure(figsize=(8,5))
    df = []
    for i,name in enumerate(label_names):
        df.append(values[labels==i])
    sns.boxplot(data=df)
    plt.xticks(range(len(label_names)), label_names, rotation=30)
    plt.title(title)
    plt.show()
```

```python
# mdl_pipeline.py
import numpy as np
import gzip, bz2, lzma
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from typing import List, Tuple

# ----------------------------
# Compression / KC helpers
# ----------------------------
def compress_bytes(data: bytes, method: str = 'gzip') -> bytes:
    if method == 'gzip':
        return gzip.compress(data, compresslevel=9)
    elif method == 'bz2':
        return bz2.compress(data, compresslevel=9)
    elif method == 'lzma':
        return lzma.compress(data, preset=9)
    else:
        raise ValueError(f"Unknown compressor: {method}")

def kolmogorov_complexity_array(arr: np.ndarray, method: str = 'gzip') -> float:
    """Normalized compressed length for a numpy array (image or feature vector)."""
    b = arr.tobytes()
    if len(b) == 0:
        return 0.0
    clen = len(compress_bytes(b, method))
    return clen / len(b)

def conditional_complexity_bytes(b1: bytes, b2: bytes, method: str = 'gzip') -> float:
    """NCD-like normalized information distance from bytes (works on pre-serialized arrays)."""
    c1 = len(compress_bytes(b1, method))
    c2 = len(compress_bytes(b2, method))
    c12 = len(compress_bytes(b1 + b2, method))
    # normalized information distance (NID)
    if max(c1, c2) == 0:
        return 0.0
    nid = (c12 - min(c1, c2)) / max(c1, c2)
    return nid

# ----------------------------
# Image-based MDL (prototypes = mean images)
# ----------------------------
def build_image_prototypes(images: np.ndarray, labels: np.ndarray) -> dict:
    """
    images: array-like shape (N, H, W) or list of 2D arrays
    returns dict: class_label -> prototype_image (uint8 float mean)
    """
    classes = np.unique(labels)
    prototypes = {}
    for c in classes:
        imgs_c = np.stack([img.astype(np.float32) for img, lbl in zip(images, labels) if lbl == c], axis=0)
        proto = imgs_c.mean(axis=0)
        # convert to stable uint8 representation for compression
        proto_u8 = _to_uint8(proto)
        prototypes[c] = proto_u8
    return prototypes

def _to_uint8(img: np.ndarray, clip_percentiles=(1, 99)) -> np.ndarray:
    """Robust scaling of float image to uint8."""
    lo, hi = np.percentile(img, clip_percentiles)
    imgc = np.clip(img, lo, hi)
    if hi == lo:
        return np.zeros_like(imgc, dtype=np.uint8)
    scaled = ((imgc - lo) / (hi - lo) * 255.0).astype(np.uint8)
    return scaled

def mdl_classifier_images(X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray,
                          compressor: str = 'gzip', verbose: bool = True) -> np.ndarray:
    """
    Pure AIT MDL classifier using image prototypes (mean images).
    For each class prototype P_c: predict class that minimizes NCD(test_image, P_c).
    This is slow if many classes or large images. Caches compressed prototypes.
    """
    classes = np.unique(y_train)
    # Build prototypes from train images
    protos = build_image_prototypes(X_train, y_train)

    # Precompute compressed bytes for prototypes
    proto_bytes = {c: protos[c].tobytes() for c in protos}
    proto_comp_len = {c: len(compress_bytes(proto_bytes[c], compressor)) for c in proto_bytes}

    preds = []
    n = len(X_test)
    for i, img in enumerate(X_test):
        if verbose and (i+1) % 50 == 0:
            print(f"[MDL-image] processing {i+1}/{n}")
        img_u8 = _to_uint8(img)
        img_b = img_u8.tobytes()
        img_c_len = len(compress_bytes(img_b, compressor))
        best_c = None
        best_score = float('inf')
        # compare to prototypes
        for c, p_b in proto_bytes.items():
            # compute cxy compress length (compress test + proto)
            cxy = len(compress_bytes(img_b + p_b, compressor))
            # compute normalized information distance
            nid = (cxy - min(img_c_len, proto_comp_len[c])) / max(img_c_len, proto_comp_len[c])
            if nid < best_score:
                best_score = nid
                best_c = c
        preds.append(best_c)
    return np.array(preds, dtype=int)

# ----------------------------
# Feature-based MDL (practical)
# ----------------------------
def mdl_classifier_features(X_train_feat: np.ndarray, y_train: np.ndarray, X_test_feat: np.ndarray,
                            distance: str = 'euclid', verbose: bool = False) -> np.ndarray:
    """
    Build prototype = mean feature vector per class and predict by nearest prototype.
    distance: 'euclid' or 'manhattan' or 'cosine' (cosine uses sklearn pairwise)
    """
    from sklearn.metrics import pairwise_distances

    classes = np.unique(y_train)
    prototypes = np.vstack([X_train_feat[y_train == c].mean(axis=0) for c in classes])  # (n_classes, D)
    # compute pairwise distances: shape (n_test, n_classes)
    D = pairwise_distances(X_test_feat, prototypes, metric='euclidean' if distance == 'euclid' else distance)
    idx = np.argmin(D, axis=1)
    preds = classes[idx]
    return preds

# ----------------------------
# Comparison utilities
# ----------------------------
def evaluate_and_print(y_true: np.ndarray, y_pred: np.ndarray, label_names: List[str] = None, title: str = ''):
    acc = accuracy_score(y_true, y_pred)
    print(f"\n=== Evaluation: {title} ===")
    print(f"Accuracy: {acc:.4f}")
    if label_names is not None:
        print(classification_report(y_true, y_pred, target_names=label_names))
    else:
        print(classification_report(y_true, y_pred))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=label_names if label_names else np.unique(y_true),
                yticklabels=label_names if label_names else np.unique(y_true))
    plt.title(title + " Confusion Matrix")
    plt.ylabel("True")
    plt.xlabel("Pred")
    plt.show()

# ----------------------------
# Baseline ML models (RandomForest, SVM) trained on features
# ----------------------------
def train_random_forest(X_train_feat: np.ndarray, y_train: np.ndarray, n_estimators: int = 200, random_state: int = 42):
    rf = RandomForestClassifier(n_estimators=n_estimators, random_state=random_state, n_jobs=-1)
    rf.fit(X_train_feat, y_train)
    return rf

def train_svm(X_train_feat: np.ndarray, y_train: np.ndarray, C: float = 1.0, kernel: str = 'rbf'):
    svm = SVC(C=C, kernel=kernel, probability=True)
    svm.fit(X_train_feat, y_train)
    return svm

# ----------------------------
# Full pipeline function
# ----------------------------
def run_mdl_vs_ml_pipeline(X_train_images: np.ndarray, y_train: np.ndarray,
                           X_test_images: np.ndarray, y_test: np.ndarray,
                           extract_features_fn,
                           compressors: List[str] = ('gzip', 'bz2', 'lzma'),
                           label_names: List[str] = None,
                           mode: str = 'features'):
    """
    mode: 'image' -> run pure image MDL (slow)
          'features' -> extract features, run MDL-on-features and ML baselines
    extract_features_fn: function(image) -> 1D array of features (used if mode == 'features')
    """
    if mode not in ('image', 'features'):
        raise ValueError("mode must be 'image' or 'features'")

    if mode == 'image':
        # Pure AIT MDL: one compressor at a time (slow). We'll use gzip by default.
        for comp in compressors:
            print(f"\n--- MDL (image) with compressor: {comp} ---")
            y_pred = mdl_classifier_images(X_train_images, y_train, X_test_images, compressor=comp, verbose=True)
            evaluate_and_print(y_test, y_pred, label_names=label_names, title=f"MDL-image ({comp})")
        return

    # ---------- features mode ----------
    # 1) Extract features for train/test
    print("[*] Extracting features (train)...")
    X_train_feat = np.vstack([extract_features_fn(img) for img in X_train_images])
    print("[*] Extracting features (test)...")
    X_test_feat = np.vstack([extract_features_fn(img) for img in X_test_images])

    # 2) Normalize features
    scaler = StandardScaler()
    X_train_feat = scaler.fit_transform(X_train_feat)
    X_test_feat = scaler.transform(X_test_feat)

    # 3) MDL-on-features
    print("\n--- MDL (features, prototype nearest) ---")
    y_pred_mdl = mdl_classifier_features(X_train_feat, y_train, X_test_feat, distance='euclid')
    evaluate_and_print(y_test, y_pred_mdl, label_names=label_names, title="MDL-features")

    # 4) Random Forest baseline
    print("\n--- Random Forest (features) ---")
    rf = train_random_forest(X_train_feat, y_train)
    y_pred_rf = rf.predict(X_test_feat)
    evaluate_and_print(y_test, y_pred_rf, label_names=label_names, title="RandomForest")

    # 5) SVM baseline
    print("\n--- SVM (features) ---")
    svm = train_svm(X_train_feat, y_train, C=1.0, kernel='rbf')
    y_pred_svm = svm.predict(X_test_feat)
    evaluate_and_print(y_test, y_pred_svm, label_names=label_names, title="SVM (RBF)")

    return {
        'X_train_feat': X_train_feat, 'X_test_feat': X_test_feat,
        'y_pred_mdl': y_pred_mdl, 'y_pred_rf': y_pred_rf, 'y_pred_svm': y_pred_svm
    }

# ----------------------------
# EXAMPLE extract_features_fn
# You should replace this with your previously-built function,
# e.g. extract_rich_features_improved or extract_rich_features_ultra
# ----------------------------
def example_extract_features(img: np.ndarray) -> np.ndarray:
    """
    Minimal example: returns a compact vector of AIT-inspired features:
    [compression(gzip), entropy, sobel_compression, mean_patch_kc, std_patch_kc]
    Replace with your richer extractor for better results.
    """
    # simple helpers (re-define or import in your project)
    from skimage.filters import sobel
    from skimage.measure import shannon_entropy
    # compression helper (reuse compress_bytes)
    img_u8 = _to_uint8(img)
    b = img_u8.tobytes()
    comp_gzip = len(compress_bytes(b, 'gzip')) / max(len(b), 1)
    ent = shannon_entropy(img_u8)
    edges = sobel(img.astype(float))
    edges_u8 = _to_uint8(edges)
    edges_comp = len(compress_bytes(edges_u8.tobytes(), 'gzip')) / max(len(edges_u8.tobytes()), 1)
    # simple patches kc stats
    # non-overlapping 32x32
    h, w = img.shape
    pcs = []
    psize = 32
    for i in range(0, h - psize + 1, psize):
        for j in range(0, w - psize + 1, psize):
            p = _to_uint8(img[i:i+psize, j:j+psize])
            pcs.append(len(compress_bytes(p.tobytes(), 'gzip')) / max(len(p.tobytes()), 1))
    if len(pcs) == 0:
        mean_p, std_p = 0.0, 0.0
    else:
        mean_p, std_p = float(np.mean(pcs)), float(np.std(pcs))
    return np.array([comp_gzip, ent, edges_comp, mean_p, std_p], dtype=float)


# ----------------------------
# Example usage (to be adapted)
# ----------------------------
if __name__ == "__main__":

    # Run MDL-on-features and compare to RF/SVM
    results = run_mdl_vs_ml_pipeline(X_train, y_train, X_test, y_test,
                                     extract_features_fn=example_extract_features,
                                     compressors=('gzip','bz2','lzma'),
                                     label_names=label_names,
                                     mode='features')

    # If you want to run pure image MDL (SLOW), set mode='image'
    # run_mdl_vs_ml_pipeline(X_train_images, y_train, X_test_images, y_test,
    #                         extract_features_fn=example_extract_features,
    #                         compressors=('gzip',),
    #                         label_names=label_names,
    #                         mode='image')
```
